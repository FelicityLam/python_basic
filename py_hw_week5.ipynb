```python
import xlrd
import pandas as pd

# Open the workbook
wb = xlrd.open_workbook('H1.xls')

# List to store hyperlinks
hyperlinks = []

# Iterate through each sheet in the workbook
for sheet in wb.sheets():
    for row in range(sheet.nrows):
        for col in range(sheet.ncols):
            cell = sheet.cell(row, col)
            if cell.ctype == xlrd.XL_CELL_TEXT:
                # Check if the cell contains a URL
                if cell.value.startswith('http://') or cell.value.startswith('https://'):
                    hyperlinks.append(cell.value)

# Create a DataFrame
df_hyperlinks = pd.DataFrame(hyperlinks, columns=['Hyperlinks'])

# Save the DataFrame to a new Excel file (.xlsx)
df_hyperlinks.to_excel('New_H1.xlsx', index=False)
```
